```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import csv
import matplotlib.tri as tri
import os, subprocess

#inputdir = '../gauss/vor_i_output_gnuplot/'
#inputdir = '../gauss/part_dx004_r012/'
inputdir = '../gauss/'
#inputdir = os.getcwd()
print(inputdir)

nx=50
ny=50
tot=nx*ny


npts = 2508
t1 = 24 #total time
m  = 6000 #total step
dt = t1/m # step size
fcount = 0
fmax   = 601


files = os.listdir(inputdir)
files.sort()


zmin=0
zmax=20
lev = np.linspace(zmin, zmax, 21)
#print(lev)

for file in files:
    #print (file)
    if file.endswith('.dat'):
        
        step   = os.path.splitext(file)[0]
        step   = int(step)
        time   = step*dt
        #print ('dt, step, time='+str(dt)+' '+str(step)+' '+str(step*dt))
        fcount = fcount+1
        if fcount > fmax:
            break
    
        x=[]
        y=[]
        z=[]

        data_path=inputdir+'/'+file
        print ('processing file :', data_path)

        
        #read data
        line_count=0
        with open(data_path, 'r') as file:
            data = csv.reader(file, delimiter=' ')
            for row in data:
                ix = int(line_count/ny)
                iy = line_count%ny
                #print(line_count, ix, iy)
                x.append(float(row[0]))
                y.append(float(row[1]))
                z.append(float(row[2]))
                line_count += 1

        fig, ax = plt.subplots()
        ax.plot(x, y, 'ko', ms=1) #show grid points
        ax.set_aspect('equal')
        ax.set(xlim=(-2.5, 2.5), ylim=(-2.5, 2.5))
        #cp = ax.contour(x, y, z, levels=lev)
        #fig.colorbar(cp) # Add a colorbar to a plot
        #ax.clabel(cp, inline=1, fontsize=10)
        title="#Particles:" + str(npts)+' dt:'+str(dt)+ ' step:'+str(step)+' time:'+str(round(time,2))
        
        ax.set_title(title)
        output='part'+str(fcount-1).zfill(6)+'.png'
        print("output to "+output+"\n")
        fig.savefig(output, dpi=300)
        plt.close()
```
